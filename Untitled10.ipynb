```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the uploaded Excel file
file_path = 'Doyang111 (1).xlsx'
data = pd.read_excel(file_path)

# Rename columns for clarity
data.columns = ['Date', 'Month', 'Year', 'DC', 'Energy', 'Waterlevel', 'Normal_Rainfall', 'Actual_Rainfall']

# Drop the first row which seems to contain headers
data = data.drop(0)

# Convert columns to appropriate data types
data = data.astype({
    'Date': int,
    'Month': int,
    'Year': int,
    'DC': float,
    'Energy': float,
    'Waterlevel': float,
    'Normal_Rainfall': float,
    'Actual_Rainfall': float
})

# Calculate the correlation matrix
correlation_matrix = data[['Actual_Rainfall', 'Waterlevel', 'DC', 'Energy']].corr()

# Plotting the heatmap for correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix Heatmap')
plt.show()

# Scatter plot matrix for all combinations
sns.pairplot(data[['Actual_Rainfall', 'Waterlevel', 'DC', 'Energy']])
plt.suptitle('Scatter Plot Matrix', y=1.02)
plt.show()

# Individual scatter plots with regression lines
sns.lmplot(x='Actual_Rainfall', y='Waterlevel', data=data, aspect=1.5)
plt.title('Actual Rainfall vs. Water Level')
plt.show()

sns.lmplot(x='Actual_Rainfall', y='DC', data=data, aspect=1.5)
plt.title('Actual Rainfall vs. DC')
plt.show()

sns.lmplot(x='Actual_Rainfall', y='Energy', data=data, aspect=1.5)
plt.title('Actual Rainfall vs. Energy')
plt.show()

sns.lmplot(x='Waterlevel', y='DC', data=data, aspect=1.5)
plt.title('Water Level vs. DC')
plt.show()

sns.lmplot(x='Waterlevel', y='Energy', data=data, aspect=1.5)
plt.title('Water Level vs. Energy')
plt.show()

sns.lmplot(x='DC', y='Energy', data=data, aspect=1.5)
plt.title('DC vs. Energy')
plt.show()
```
