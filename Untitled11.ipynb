```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the uploaded Excel file
file_path = 'Doyang111 (1).xlsx'
data = pd.read_excel(file_path)

# Rename columns for clarity
data.columns = ['Date', 'Month', 'Year', 'DC', 'Energy', 'Waterlevel', 'Normal_Rainfall', 'Actual_Rainfall']

# Drop the first row which seems to contain headers
data = data.drop(0)

# Convert columns to appropriate data types
data = data.astype({
    'Date': int,
    'Month': int,
    'Year': int,
    'DC': float,
    'Energy': float,
    'Waterlevel': float,
    'Normal_Rainfall': float,
    'Actual_Rainfall': float
})

# Calculate the correlation matrix
correlation_matrix = data[['Actual_Rainfall', 'Waterlevel', 'DC', 'Energy']].corr()

# Plotting the heatmap for correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix Heatmap')
plt.show()

# Scatter plot matrix for all combinations
sns.pairplot(data[['Actual_Rainfall', 'Waterlevel', 'DC', 'Energy']])
plt.suptitle('Scatter Plot Matrix', y=1.02)
plt.show()

# Individual scatter plots with regression lines
sns.lmplot(x='Actual_Rainfall', y='Waterlevel', data=data, aspect=1.5)
plt.title('Actual Rainfall vs. Water Level')
plt.show()

sns.lmplot(x='Actual_Rainfall', y='DC', data=data, aspect=1.5)
plt.title('Actual Rainfall vs. DC')
plt.show()

sns.lmplot(x='Actual_Rainfall', y='Energy', data=data, aspect=1.5)
plt.title('Actual Rainfall vs. Energy')
plt.show()

sns.lmplot(x='Waterlevel', y='DC', data=data, aspect=1.5)
plt.title('Water Level vs. DC')
plt.show()

sns.lmplot(x='Waterlevel', y='Energy', data=data, aspect=1.5)
plt.title('Water Level vs. Energy')
plt.show()

sns.lmplot(x='DC', y='Energy', data=data, aspect=1.5)
plt.title('DC vs. Energy')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Excel file
file_path = 'Lotak.xlsx'
data = pd.read_excel(file_path)

# Rename columns for easier access
data.columns = ['Date', 'Month', 'Year', 'DC', 'Energy', 'Water_Level', 'Bishnupur_Normal', 'Bishnupur_Actual',
                'Imphal_East_Normal', 'Imphal_East_Actual', 'Thoubal_Normal', 'Thoubal_Actual']

# Extract relevant columns
relevant_data = data[['Date', 'Month', 'Year', 'DC', 'Energy', 'Water_Level', 'Bishnupur_Actual',
                      'Imphal_East_Actual', 'Thoubal_Actual']]

# Drop the first row of headers
relevant_data = relevant_data.drop(0)

# Convert columns to numeric where applicable
relevant_data = relevant_data.apply(pd.to_numeric, errors='coerce')

# Drop rows with any NaN values to ensure clean data for correlation
relevant_data = relevant_data.dropna()

# Calculate the correlation matrix
correlation_matrix = relevant_data.corr()

# Plot the heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()

# Function to annotate correlation coefficients on pair plots
def annotate_correlation(data, **kws):
    r = data.corr().iloc[0, 1]
    ax = plt.gca()
    ax.set_title(f'r = {r:.2f}', fontsize=10)

# Plot pair plots with regression lines and correlation coefficients
sns.pairplot(relevant_data, kind='reg', plot_kws={'line_kws':{'color':'red'}, 'scatter_kws': {'alpha': 0.5}})
plt.show()
```
