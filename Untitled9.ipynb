```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel files
file_path_water = 'Water_level_Data.xlsx'
file_path_rainfall = 'Rainfall data.xlsx'

# Load the data from the relevant sheets
df_water = pd.read_excel(file_path_water, sheet_name='2023', skiprows=2)
df_rainfall = pd.read_excel(file_path_rainfall, sheet_name='csvFile207', skiprows=2)

# Rename the columns for clarity in water data
df_water.columns = ['Index', 'Date', 'Doyang_Water_Level', 'Ranganadi_Water_Level', 'Loktak_Water_Level',
                    'Unknown1', 'Doyang_DC', 'Doyang_Energy', 'Unknown2', 'Ranganadi_DC', 'Unknown3',
                    'Loktak_DC', 'Loktak_Energy']

# Drop the unnecessary columns
df_water = df_water.drop(columns=['Index', 'Unknown1', 'Unknown2', 'Unknown3'])

# Convert the 'Date' column to datetime format
df_water['Date'] = pd.to_datetime(df_water['Date'])

# Rename the columns for clarity in rainfall data
df_rainfall.columns = ['Date', 'Loktak_Normal_mm', 'Loktak_Actual_mm', 'Unknown1', 'Unknown2',
                       'Imphal_East_Normal_mm', 'Imphal_East_Actual_mm', 'Unknown3', 'Unknown4',
                       'Thoubal_Normal_mm', 'Thoubal_Actual_mm', 'Unknown5', 'Unknown6', 'Unknown7',
                       'Unknown8', 'Unknown9', 'Unknown10', 'Unknown11', 'East_Kameng_Normal_mm',
                       'East_Kameng_Actual_mm', 'Unknown12', 'Kurung_Kumey_Normal_mm', 'Kurung_Kumey_Actual_mm',
                       'Unknown13', 'Unknown14', 'Doyang_Normal_mm', 'Doyang_Actual_mm']

# Drop the unnecessary columns in rainfall data
df_rainfall = df_rainfall.drop(columns=['Unknown1', 'Unknown2', 'Unknown3', 'Unknown4', 'Unknown5', 'Unknown6',
                                        'Unknown7', 'Unknown8', 'Unknown9', 'Unknown10', 'Unknown11', 'Unknown12',
                                        'Unknown13', 'Unknown14'])

# Convert the 'Date' column to datetime format
df_rainfall['Date'] = pd.to_datetime(df_rainfall['Date'], errors='coerce')

# Merge the dataframes on 'Date'
df_merged = pd.merge(df_water, df_rainfall, on='Date', how='left')

# Investigate NaN values
print("NaN values in each column:\n", df_merged.isna().sum())

# Calculate correlations
correlations = df_merged.corr()

# Display the correlations
print("Correlations:\n", correlations)

# Visualize the correlations using a heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(correlations, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Scatter plots to visualize relationships
fig, axes = plt.subplots(3, 3, figsize=(15, 15))

# Scatter plots of rainfall vs water levels, DC, and energy
sns.scatterplot(ax=axes[0, 0], x='Loktak_Actual_mm', y='Loktak_Water_Level', data=df_merged)
axes[0, 0].set_title('Loktak Rainfall vs Water Level')

sns.scatterplot(ax=axes[0, 1], x='Loktak_Actual_mm', y='Loktak_DC', data=df_merged)
axes[0, 1].set_title('Loktak Rainfall vs DC')

sns.scatterplot(ax=axes[0, 2], x='Loktak_Actual_mm', y='Loktak_Energy', data=df_merged)
axes[0, 2].set_title('Loktak Rainfall vs Energy')

sns.scatterplot(ax=axes[1, 0], x='Doyang_Actual_mm', y='Doyang_Water_Level', data=df_merged)
axes[1, 0].set_title('Doyang Rainfall vs Water Level')

sns.scatterplot(ax=axes[1, 1], x='Doyang_Actual_mm', y='Doyang_DC', data=df_merged)
axes[1, 1].set_title('Doyang Rainfall vs DC')

sns.scatterplot(ax=axes[1, 2], x='Doyang_Actual_mm', y='Doyang_Energy', data=df_merged)
axes[1, 2].set_title('Doyang Rainfall vs Energy')

sns.scatterplot(ax=axes[2, 0], x='East_Kameng_Actual_mm', y='Ranganadi_Water_Level', data=df_merged)
axes[2, 0].set_title('Ranganadi Rainfall vs Water Level')

sns.scatterplot(ax=axes[2, 1], x='East_Kameng_Actual_mm', y='Ranganadi_DC', data=df_merged)
axes[2, 1].set_title('Ranganadi Rainfall vs DC')

plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load and clean the data (same as before)
file_path_water = 'Water_level_Data.xlsx'
file_path_rainfall = 'Rainfall data.xlsx'
df_water = pd.read_excel(file_path_water, sheet_name='2023', skiprows=2)
df_rainfall = pd.read_excel(file_path_rainfall, sheet_name='csvFile207', skiprows=2)
df_water.columns = ['Index', 'Date', 'Doyang_Water_Level', 'Ranganadi_Water_Level', 'Loktak_Water_Level', 'Unknown1', 'Doyang_DC', 'Doyang_Energy', 'Unknown2', 'Ranganadi_DC', 'Unknown3', 'Loktak_DC', 'Loktak_Energy']
df_water = df_water.drop(columns=['Index', 'Unknown1', 'Unknown2', 'Unknown3'])
df_water['Date'] = pd.to_datetime(df_water['Date'])
df_rainfall.columns = ['Date', 'Loktak_Normal_mm', 'Loktak_Actual_mm', 'Unknown1', 'Unknown2', 'Imphal_East_Normal_mm', 'Imphal_East_Actual_mm', 'Unknown3', 'Unknown4', 'Thoubal_Normal_mm', 'Thoubal_Actual_mm', 'Unknown5', 'Unknown6', 'Unknown7', 'Unknown8', 'Unknown9', 'Unknown10', 'Unknown11', 'East_Kameng_Normal_mm', 'East_Kameng_Actual_mm', 'Unknown12', 'Kurung_Kumey_Normal_mm', 'Kurung_Kumey_Actual_mm', 'Unknown13', 'Unknown14', 'Doyang_Normal_mm', 'Doyang_Actual_mm']
df_rainfall = df_rainfall.drop(columns=['Unknown1', 'Unknown2', 'Unknown3', 'Unknown4', 'Unknown5', 'Unknown6', 'Unknown7', 'Unknown8', 'Unknown9', 'Unknown10', 'Unknown11', 'Unknown12', 'Unknown13', 'Unknown14'])
df_rainfall['Date'] = pd.to_datetime(df_rainfall['Date'], errors='coerce')
df_merged = pd.merge(df_water, df_rainfall, on='Date', how='left')

# Perform simple linear regression to predict Loktak Water Level from Loktak Rainfall
X = df_merged[['Loktak_Actual_mm']].dropna()  # Independent variable
y = df_merged['Loktak_Water_Level'].dropna()  # Dependent variable

# Ensure X and y have the same length
common_index = X.index.intersection(y.index)
X = X.loc[common_index]
y = y.loc[common_index]

# Fit the model
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Print the coefficients
print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)
print('Mean squared error (MSE): %.2f' % mean_squared_error(y, y_pred))
print('Coefficient of determination (R^2): %.2f' % r2_score(y, y_pred))

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression Line')
plt.xlabel('Loktak Actual Rainfall (mm)')
plt.ylabel('Loktak Water Level')
plt.title('Simple Linear Regression: Loktak Rainfall vs Water Level')
plt.legend()
plt.show()
```
